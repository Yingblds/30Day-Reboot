```python
import pandas as pd

# Load sample dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

# Inspect
print("First 5 rows:")
print(df.head())

print("\nDataset info:")  # \n -- Add a blank line before the output
print(df.info())

print("\nDataset description:")
print(df.describe())

```
